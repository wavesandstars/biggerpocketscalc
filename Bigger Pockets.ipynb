```python
#blue print for taking in different people's real estate investment 
#take input for different attributes 


class BiggerPockets():
    def __init__(self, expenses=0, cashflow=0, cashoncash=0, income=0):
        self.income = income
        self.expenses= expenses
        self.cashflow = cashflow
        self.cashoncash = cashoncash
        
        
        
    def getIncome(self):
        rent = input('How much do you make from rent?')
        otherincome = input('How much do you collect from other rental charges?')
        self.income = int(rent) + int(otherincome)
        #totalincome = (int(rent) + int(otherincome))
        #self.income.append(totalincome)
        
    def showIncome(self):
        print(self.income)
#         if len(self.income)== 0:
#             print('no income entered')
#         else:
#             print(self.income)
    
    def getExpenses(self):
        taxes = input('What do you pay in taxes?')
        utilities = input('What do you pay in utilites?')
        hoa = input('What do you pay in HOA fees?')
        repairs = input('What will you set aside for repairs?')
        cap_ex = input('What is your cap ex?')
        prop_man = input('What do you pay for property managment?')
        mortgage = input('What do you pay for your mortgage?')
        self.expenses = (int(taxes)+int(utilities)+int(hoa)+int(repairs)+int(cap_ex)+int(prop_man)+int(mortgage))
        #self.expenses.append(totalexpenses)
       
    def showExpenses(self):
        print(self.expenses)
        
#         if len(self.expenses)== 0:
#             print('no expenses entered')
#         else:
#             print(self.expenses)
               
            
    def cashFlow(self):
        self.cashflow = int(self.income)- int(self.expenses)
        print(self.cashflow)
    
    def cc(self):
        downpayment = input('What was your down payment?')
        closing = input('What was your closing costs?')
        repairmoney = input('How much money was used for repairs?')
        misc= input('How much did you pay for misc costs?')
        self.cashoncash = int(downpayment) +int(closing) +int(repairmoney) +int(misc)
        #self.cashoncash.append(totalinvestment)
      
 
    def showcc(self):
        print(self.cashoncash)

    
    def annualCash(self):
        annualcash = ((int(self.income)-int(self.expenses)) * 12)
        print(annualcash)
        roi = ((annualcash) / (self.cashoncash))
        print((f'Your roi is {(roi * 100)}%') )
    
   

def run():
    person = BiggerPockets()
    while True:
        response = input('What would you like to do: add income, add expenses, see income, see expenses, see cash flow, cash on cash, find roi, or quit?')
        if response.lower() == 'add income':
            person.getIncome()
        if response.lower() == 'see income':
            person.showIncome()
        if response.lower() == 'add expenses':
            person.getExpenses()
        if response.lower()== 'see expenses':
            person.showExpenses()
        if response.lower() == 'see cash flow':
            person.cashFlow()
        if response.lower()== 'cash on cash':
            person.cc()
        if response.lower() == 'find roi':
            person.annualCash()
        if response.lower() == 'quit':
            break
    
run()

 





        

        

        
       
```
